```python
top_dict = {}
```

```python
for i in df['text']:
  ii = i.split(" ")
  for iii in ii:
    if iii in top_dict:
      top_dict[iii] += 1
    else:
      top_dict[iii] = 1
```

```python
def show_wordcloud(data):
    WC = WordCloud(
        font_path='NotoSansEthiopic-Black.ttf',
        background_color='white',
        #stopwords=stopwords,
        #max_words=1000,
        max_font_size=500, 
        scale=5,
        random_state=1 
    )
    wordcloud = WC.generate_from_frequencies(data)

    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')
    
    plt.imshow(wordcloud)
    plt.show()

show_wordcloud(top_dict)
```
